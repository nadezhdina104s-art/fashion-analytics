```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.family'] = 'DejaVu Sans'

print("=" * 80)
print("A/B ТЕСТ: ДИНАМИЧЕСКИЕ СКИДКИ vs ФИКСИРОВАННЫЕ 70%")
print("=" * 80)

# ============================================
# 1. СОЗДАНИЕ РЕАЛИСТИЧНЫХ ДАННЫХ
# ============================================
print("\n🧪 Подготовка данных для A/B теста...")

np.random.seed(42)
n_products = 200

# Создаём товары с реалистичными параметрами
recommendations = pd.DataFrame({
    'article_id': range(1, n_products + 1),
    'product_type': np.random.choice(['Dress', 'Blouse', 'Trousers', 'Jeans', 'Jacket', 'Skirt'], n_products),
    'current_stock': np.random.randint(20, 150, n_products),
    'full_price': np.random.uniform(2000, 10000, n_products),
    'optimal_discount_pct': np.random.uniform(20, 65, n_products),  # Оптимальные скидки 20-65%
})
# Себестоимость = 40-60% от цены
recommendations['cost'] = recommendations['full_price'] * np.random.uniform(0.4, 0.6, n_products)

print(f"✅ Создано {n_products} товаров")
print(f"   • Средняя цена: {recommendations['full_price'].mean():.0f} руб")
print(f"   • Средняя себестоимость: {recommendations['cost'].mean():.0f} руб")
print(f"   • Средняя оптимальная скидка: {recommendations['optimal_discount_pct'].mean():.0f}%")

# ============================================
# 2. РАЗДЕЛЕНИЕ НА ГРУППЫ
# ============================================
# Стратифицированное разделение
recommendations['stock_level'] = pd.cut(
    recommendations['current_stock'],
    bins=3,
    labels=['low', 'med', 'high']
).astype(str)

recommendations['stratum'] = recommendations['product_type'].astype(str) + '_' + recommendations['stock_level']

control_idx, test_idx = train_test_split(
    recommendations.index,
    test_size=0.5,
    stratify=recommendations['stratum'],
    random_state=42
)

recommendations['group'] = 'Control'
recommendations.loc[test_idx, 'group'] = 'Test'

print(f"\n📊 Разделение на группы:")
print(f"   • Контрольная (A): {len(control_idx)} товаров (фикс. 70%)")
print(f"   • Тестовая (B): {len(test_idx)} товаров (оптимальные скидки)")

# ============================================
# 3. ФУНКЦИЯ ПРИБЫЛИ (ИСПРАВЛЕНО)
# ============================================
def simulate_profit_realistic(row, discount):
    """
    Реалистичная симуляция прибыли при заданной скидке
    """
    price = row['full_price'] * (1 - discount/100)
    cost = row['cost']
    stock = row['current_stock']

    # Если цена ниже себестоимости + 10% маржа, продаём только если очень нужно
    if price < cost * 1.1:
        # Продаём только 20% остатка (остальное - убыток)
        base_demand = stock * 0.2
    else:
        # Нормальная маржа - спрос зависит от скидки
        base_demand = stock * 0.3
        # Скидка увеличивает спрос
        demand_multiplier = 1 + (discount / 100) * 1.5
        base_demand = base_demand * demand_multiplier

    # Добавляем случайность (покупательское поведение)
    sold = min(stock, base_demand * np.random.uniform(0.8, 1.2))
    sold = int(sold)

    # Прибыль = выручка - себестоимость проданного
    # + ликвидационная стоимость остатков (10% от себестоимости)
    revenue = sold * price
    total_cost = sold * cost
    leftover = stock - sold
    salvage = leftover * cost * 0.1

    profit = revenue - total_cost + salvage
    return max(-stock * cost * 0.1, profit)  # Не даём уйти в глубокий минус

# ============================================
# 4. СИМУЛЯЦИЯ РЕЗУЛЬТАТОВ
# ============================================
print("\n🔄 Симуляция результатов...")

# Контрольная группа (70% скидка)
control_results = []
for _, row in recommendations[recommendations['group'] == 'Control'].iterrows():
    profit = simulate_profit_realistic(row, 70)
    control_results.append(profit)

# Тестовая группа (оптимальные скидки)
test_results = []
test_discounts = []
for _, row in recommendations[recommendations['group'] == 'Test'].iterrows():
    optimal_discount = row['optimal_discount_pct']
    profit = simulate_profit_realistic(row, optimal_discount)
    test_results.append(profit)
    test_discounts.append(optimal_discount)

print(f"   • Контроль (70%): средняя прибыль = {np.mean(control_results):,.0f} руб")
print(f"   • Тест (оптим.): средняя прибыль = {np.mean(test_results):,.0f} руб")
print(f"   • Средняя скидка в тесте: {np.mean(test_discounts):.0f}%")

# ============================================
# 5. СТАТИСТИЧЕСКИЙ ТЕСТ
# ============================================
print("\n" + "=" * 60)
print("РЕЗУЛЬТАТЫ A/B ТЕСТА")
print("=" * 60)

print(f"\n📊 Описательная статистика:")
print(f"   • Контроль (A): среднее = {np.mean(control_results):,.0f} руб, std = {np.std(control_results):,.0f}")
print(f"   • Тест (B):    среднее = {np.mean(test_results):,.0f} руб, std = {np.std(test_results):,.0f}")

# t-test
t_stat, p_value = stats.ttest_ind(test_results, control_results, alternative='greater')

print(f"\n🔬 Статистический тест (t-test):")
print(f"   • t-статистика: {t_stat:.4f}")
print(f"   • p-value: {p_value:.6f}")

alpha = 0.05
if p_value < alpha:
    print(f"   ✅ p-value < 0.05 → РАЗЛИЧИЕ СТАТИСТИЧЕСКИ ЗНАЧИМО!")
    print(f"   ✅ Тестовая группа ПОБЕДИЛА!")
else:
    print(f"   ❌ p-value >= 0.05 → различие НЕ значимо")

# Размер эффекта
pooled_std = np.sqrt((np.std(control_results, ddof=1)**2 + np.std(test_results, ddof=1)**2) / 2)
cohens_d = (np.mean(test_results) - np.mean(control_results)) / pooled_std

print(f"\n📏 Размер эффекта (Cohen's d): {cohens_d:.3f}")
if abs(cohens_d) < 0.2:
    effect_size = "малый"
elif abs(cohens_d) < 0.5:
    effect_size = "средний"
elif abs(cohens_d) < 0.8:
    effect_size = "большой"
else:
    effect_size = "очень большой"
print(f"   • Интерпретация: {effect_size} эффект")

# Относительный прирост (защита от деления на ноль)
mean_control = np.mean(control_results)
if mean_control > 0:
    relative_lift = (np.mean(test_results) - mean_control) / mean_control * 100
else:
    relative_lift = (np.mean(test_results) - mean_control) / abs(mean_control) * 100 if mean_control != 0 else 100

print(f"\n💰 Относительный прирост: {relative_lift:+.1f}%")

# 95% доверительный интервал
n1, n2 = len(control_results), len(test_results)
se = np.sqrt(np.var(control_results, ddof=1)/n1 + np.var(test_results, ddof=1)/n2)
diff_mean = np.mean(test_results) - np.mean(control_results)
ci = stats.t.interval(0.95, df=n1+n2-2, loc=diff_mean, scale=se)

print(f"\n📊 95% доверительный интервал для разницы:")
print(f"   • [{ci[0]:,.0f}, {ci[1]:,.0f}] руб")

# Дополнительная метрика: ROI
control_revenue = sum(control_results)
test_revenue = sum(test_results)
control_cost = sum([row['cost'] * row['current_stock'] for _, row in recommendations[recommendations['group'] == 'Control'].iterrows()])
test_cost = sum([row['cost'] * row['current_stock'] for _, row in recommendations[recommendations['group'] == 'Test'].iterrows()])

control_roi = (control_revenue / control_cost - 1) * 100 if control_cost > 0 else 0
test_roi = (test_revenue / test_cost - 1) * 100 if test_cost > 0 else 0

print(f"\n📊 ROI (Return on Investment):")
print(f"   • Контроль: {control_roi:.1f}%")
print(f"   • Тест:    {test_roi:.1f}%")

# ============================================
# 6. ВИЗУАЛИЗАЦИЯ
# ============================================
fig = plt.figure(figsize=(18, 10))

# 6.1 Распределение прибыли
ax1 = plt.subplot(2, 3, 1)
ax1.hist(control_results, bins=20, alpha=0.6, label=f'Контроль (70%)', color='salmon', edgecolor='black')
ax1.hist(test_results, bins=20, alpha=0.6, label=f'Тест ({np.mean(test_discounts):.0f}%)', color='lightgreen', edgecolor='black')
ax1.axvline(x=np.mean(control_results), color='red', linestyle='--', linewidth=2)
ax1.axvline(x=np.mean(test_results), color='green', linestyle='--', linewidth=2)
ax1.set_xlabel('Прибыль (руб)', fontsize=12)
ax1.set_ylabel('Количество товаров', fontsize=12)
ax1.set_title('Распределение прибыли по группам', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 6.2 Boxplot
ax2 = plt.subplot(2, 3, 2)
bp = ax2.boxplot([control_results, test_results],
                  tick_labels=['Контроль (70%)', f'Тест ({np.mean(test_discounts):.0f}%)'],
                  patch_artist=True)
bp['boxes'][0].set_facecolor('salmon')
bp['boxes'][1].set_facecolor('lightgreen')
ax2.set_ylabel('Прибыль (руб)', fontsize=12)
ax2.set_title('Boxplot: сравнение распределений', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')

# 6.3 Сравнение средних с CI
ax3 = plt.subplot(2, 3, 3)
means = [np.mean(control_results), np.mean(test_results)]
errors = [1.96 * np.std(control_results, ddof=1)/np.sqrt(n1),
          1.96 * np.std(test_results, ddof=1)/np.sqrt(n2)]
bars = ax3.bar(['Контроль (70%)', f'Тест ({np.mean(test_discounts):.0f}%)'], means,
               yerr=errors, capsize=10, color=['salmon', 'lightgreen'], edgecolor='black')
ax3.set_ylabel('Средняя прибыль (руб)', fontsize=12)
ax3.set_title('Сравнение средних с 95% CI', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3, axis='y')

# 6.4 Скидки в тестовой группе
ax4 = plt.subplot(2, 3, 4)
ax4.hist(test_discounts, bins=15, color='lightblue', edgecolor='black', alpha=0.7)
ax4.axvline(x=70, color='red', linestyle='--', linewidth=2, label='Контроль (70%)')
ax4.axvline(x=np.mean(test_discounts), color='blue', linestyle='--', linewidth=2, label=f'Среднее ({np.mean(test_discounts):.0f}%)')
ax4.set_xlabel('Скидка (%)', fontsize=12)
ax4.set_ylabel('Количество товаров', fontsize=12)
ax4.set_title('Распределение скидок в тестовой группе', fontsize=14, fontweight='bold')
ax4.legend()
ax4.grid(True, alpha=0.3)

# 6.5 Прибыль vs Скидка
ax5 = plt.subplot(2, 3, 5)
# Контрольная группа
ax5.scatter([70] * len(control_results), control_results, alpha=0.5, c='salmon', label='Контроль', s=30)
# Тестовая группа
scatter = ax5.scatter(test_discounts, test_results, alpha=0.5, c='lightgreen', label='Тест', s=30)
ax5.set_xlabel('Скидка (%)', fontsize=12)
ax5.set_ylabel('Прибыль (руб)', fontsize=12)
ax5.set_title('Прибыль vs Скидка', fontsize=14, fontweight='bold')
ax5.axvline(x=70, color='red', linestyle='--', alpha=0.5)
ax5.legend()
ax5.grid(True, alpha=0.3)

# 6.6 Выводы
ax6 = plt.subplot(2, 3, 6)
ax6.axis('off')

conclusion = "✅ ГИПОТЕЗА ПОДТВЕРЖДЕНА" if p_value < 0.05 else "❌ ГИПОТЕЗА НЕ ПОДТВЕРЖДЕНА"
decision = "ЗАПУСКАТЬ на всех товарах" if p_value < 0.05 else "Требуется дополнительное тестирование"

summary_text = f"""
================================================================
                    A/B ТЕСТ: ИТОГИ
================================================================

 Гипотеза: Динамические скидки > Фикс. 70%

 Результат: {conclusion}

 Метрики:
   • Контроль (70%):     {np.mean(control_results):>10,.0f} руб
   • Тест ({np.mean(test_discounts):.0f}%):        {np.mean(test_results):>10,.0f} руб
   • Прирост:            {relative_lift:>+10.1f}%

 Статистика:
   • p-value:            {p_value:.6f}
   • Cohen's d:          {cohens_d:.3f} ({effect_size})
   • 95% CI:             [{ci[0]:,.0f}, {ci[1]:,.0f}]

 ROI:
   • Контроль:           {control_roi:.1f}%
   • Тест:               {test_roi:.1f}%

================================================================
 РЕШЕНИЕ: {decision}
================================================================
"""

ax6.text(0.05, 0.95, summary_text, transform=ax6.transAxes,
         fontsize=10, verticalalignment='top', fontfamily='monospace',
         bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9))

plt.tight_layout()
plt.savefig('ab_test_results.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✅ Визуализация сохранена: ab_test_results.png")

# ============================================
# 7. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ
# ============================================
results_df = pd.DataFrame({
    'group': ['Control', 'Test'],
    'n_products': [len(control_results), len(test_results)],
    'mean_profit': [np.mean(control_results), np.mean(test_results)],
    'std_profit': [np.std(control_results, ddof=1), np.std(test_results, ddof=1)],
    'total_profit': [sum(control_results), sum(test_results)],
    'roi': [control_roi, test_roi],
    'avg_discount': [70.0, np.mean(test_discounts)]
})
results_df.to_csv('ab_test_summary.csv', index=False, encoding='utf-8-sig', sep=';', decimal=',')
print("\n✅ Результаты сохранены: ab_test_summary.csv")

print("\n" + "=" * 80)
print("A/B ТЕСТ ЗАВЕРШЕН!")
print("=" * 80)

print(f"""
📊 ИТОГОВЫЙ ВЫВОД:

Контрольная группа (скидка 70%):
   • Средняя прибыль: {np.mean(control_results):,.0f} руб
   • ROI: {control_roi:.1f}%

Тестовая группа (оптимальные скидки {np.mean(test_discounts):.0f}%):
   • Средняя прибыль: {np.mean(test_results):,.0f} руб
   • ROI: {test_roi:.1f}%

Эффект от внедрения:
   • Дополнительная прибыль: {np.mean(test_results) - np.mean(control_results):,.0f} руб/товар
   • При масштабировании на 5000 товаров: {(np.mean(test_results) - np.mean(control_results)) * 5000:,.0f} руб
   • Улучшение ROI: {test_roi - control_roi:+.1f} п.п.
""")
```
